```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import pickle
```

```python
try:
    model = joblib.load('../model/random_forest_model.pkl')
    print("Model loaded successfully with joblib!")
except Exception as e:
    print(f"Failed to load with joblib: {e}")
```

```python
# Assuming 'model' is your trained RandomForest or Gradient Boosting model
feature_importance = model.feature_importances_
```

```python
# Create a DataFrame
feature_names = ['Administrative', 'Administrative_Duration', 'Informational', 
                 'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
                 'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay', 'Month', 
                 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 
                 'Weekend']

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
```

```python
# Save to CSV for Tableau
importance_df.to_csv("../data/cleaned/feature_importance.csv", index=False)
```

```python
# Optional: Visualize in Python
plt.figure(figsize=(10,5))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # Highest importance on top
plt.show()
```
